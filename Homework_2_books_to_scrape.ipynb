Scraping book names with CSS selector

```python
def book_names_scraper1(url):
    response = requests.get(url)
    response = TextResponse(url=response.url, body=response.text, encoding="utf-8")
    book_name1 = response.css('a[title]::text').extract()
    return book_name1
```

```python
all_book_names1 = []
for i in all_urls:
    all_book_names1.extend(book_names_scraper1(i))
    print(all_book_names1)
    time.sleep(2)
```

Scraping book names with Xpath selector

```python
def book_names_scraper2(url):
    response = requests.get(url)
    response = TextResponse(url=response.url, body=response.text, encoding="utf-8")
    book_name2 = response.xpath('//a[@title]/text()').extract()
    return book_name2
```

```python
all_book_names2 = []
for i in all_urls:
    all_book_names2.extend(book_names_scraper2(i))
    print(all_book_names2)
    time.sleep(2)
```

```python
def book_price_scraper1(url):
    response = requests.get(url)
    response = TextResponse(url=response.url, body=response.text, encoding="utf-8")
    book_price1 = response.css('p[class = "price_color"]::text').extract()
    return book_price1
```
